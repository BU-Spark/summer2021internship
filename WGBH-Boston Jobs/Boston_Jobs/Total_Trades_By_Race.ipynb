```python
import pandas as pd
```

```python
df = pd.read_csv('Copy_of_wgbh.csv')
df.head()
```

```python
df['Trade'].unique()
```

```python
df['Race_Desc'].unique()
```

```python
summary = pd.DataFrame(columns = ['TRADE', 'CAUCASIAN', 'BLACK', 'HISPANIC', 'ASIAN', 'OTHER', 'TOTAL_PROJECTS'])

trades_list = df['Trade'].unique()

trade_groups = df.groupby(df.Trade)

race_list = df['Race_Desc'].unique()

# compute percentage of employees by race
for i in range(len(trades_list)):
    temp = trade_groups.get_group(trades_list[i])
    entry = []
    total = 0
    for n in range(len(race_list)):
        filt = temp.Race_Desc == race_list[n]
        entry.append(temp[filt].shape[0])
        total += temp[filt].shape[0]
    percent = ["{:.0%}".format(int(x) / total) for x in entry]
    percent.insert(0,trades_list[i])
    percent.append(total)
    summary.loc[i] = percent
```

```python
summary
```

```python
summary.to_csv('PercentageOfMinoritiesInTrades.csv')
```

```python
dates = df['DateEntered']
for row in range(dates.shape[0]):
    # remove unnecessary timestamp
    df.at[row,'DateEntered'] = dates.iloc[row][:len(dates.iloc[row])-4]
```

```python
df
```

```python
df = df.drop(labels=0, axis=0)

date = pd.DataFrame(columns = ['Date', 'Month', 'Day', 'Year'])
date['Date'] = pd.to_datetime(df['DateEntered'])
date['Year'] = pd.DatetimeIndex(date['Date']).year
date['Month'] = pd.DatetimeIndex(date['Date']).month
date['Day'] = pd.DatetimeIndex(date['Date']).day
```

```python
df['Year'] = date['Year']
df['Year'].unique()
```

```python
summary2 = pd.DataFrame(columns = ['YEAR', 'CAUCASIAN', 'AFRICAN_AMERICAN', 'HISPANIC', 'ASIAN', 'OTHER', 'TOTAL'],
                       index = [2016, 2017, 2018, 2019, 2020, 2021])

years = df['Year'].unique()

# compute workforce composition by race for each year
for i in range(len(years)):
    filt = df['Year'] == years[i]
    temp = df[filt]
    entry = [years[i]]
    total = 0
    for n in range(len(race_list)):
        filt2 = temp.Race_Desc == race_list[n]
        entry.append(temp[filt2].shape[0])
        total += temp[filt2].shape[0]
    entry.append(total)
    summary2.loc[years[i]] = entry
```

```python
summary2.to_csv('NumberOfJobsToMinorities_PerYear.csv')
```
