## Separates apprentices and journeymen into two separate CSVs for craft level analysis

```python
import pandas as pd
import numpy
import matplotlib.pyplot as plt
```

```python
years = ['2017']
months = ['08', '10', '11'] # The only months of 2017 data that we have parsed

journeymen = pd.DataFrame() # Creating the empty Jounrymen and Apprentice DataFrames that 
                            # will hold all records from all years
apprentice = pd.DataFrame()

for y in years:
    for m in months:
        filename = '2017-18data/WorkforceUtilizationSummaryReport2017-2018' + y + m +'.csv' # Creates the filepath
                    # to each specific file based on year and month
        df = pd.read_csv(filename)
        j_filter = df.CRAFT_LEVEL == 'Journeymen' # Filter out only the Journeymen rows
        journeymen = pd.concat([journeymen, df[j_filter]]) # Add the rows to the rest of the Journeymen entries
        a_filter = df.CRAFT_LEVEL == 'Apprentice' # Filter out only the Apprentice rows
        apprentice = pd.concat([apprentice, df[a_filter]]) # Add the rows to the rest of the Apprentice entries
```

```python
years = ['2018']
months = ['01', '02', '05', '06', '08', '10', '11', '12'] # The only months of 2018 data that we have parsed

for y in years:
    for m in months:
        filename = '2017-18data/WorkforceUtilizationSummaryReport2017-2018' + y + m +'.csv'
        df = pd.read_csv(filename)
        j_filter = df.CRAFT_LEVEL == 'Journeymen'
        journeymen = pd.concat([journeymen, df[j_filter]]) # Add the rows to the rest of the Journeymen entries
        a_filter = df.CRAFT_LEVEL == 'Apprentice'
        apprentice = pd.concat([apprentice, df[a_filter]]) # Add the rows to the rest of the Apprentice entries
```

```python
years = ['2019', '2020']
months = ['01', '02', '03', '04', '05', '06', '07', '08', '09', '10', '11', '12']
# 2019 and 2020 have full years of data so we can parse through every month

for y in years:
    for m in months:
        filename = y + 'data/WorkforceUtilizationSummaryReport' + y + m +'.csv'
        df = pd.read_csv(filename)
        j_filter = df.CRAFT_LEVEL == 'Journeymen'
        journeymen = pd.concat([journeymen, df[j_filter]]) # Add the rows to the rest of the Journeymen entries
        a_filter = df.CRAFT_LEVEL == 'Apprentice'
        apprentice = pd.concat([apprentice, df[a_filter]]) # Add the rows to the rest of the Apprentice entries
```
